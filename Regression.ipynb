```python
ghfhgfch
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Data for XP and corresponding levels (50 to 74)
levels = np.arange(50, 75)
XP = np.array([2592, 2688, 2688, 2688, 2688, 2880, 2880, 2880, 3072, 3072,
               3072, 3264, 3264, 3264, 3360, 3360, 3360, 3456, 3456, 3456, 3456,
               3552, 3552, 3648, 3648])  # Add the XP value for level 74 (3648 XP)

# Rest of the code remains the same...

# Continue with the rest of the code for regression and plotting as shown before
# Reshape the data for linear regression
X = levels.reshape(-1, 1)

# Create and fit the linear regression model
reg_model = LinearRegression()
reg_model.fit(X, XP)

# Get the coefficients
a = reg_model.coef_[0]
b = reg_model.intercept_

# Generate the predicted XP values using the regression model for levels 50 to 800
all_levels = np.arange(50, 801)
predicted_XP_all = reg_model.predict(all_levels.reshape(-1, 1))

# Plot the original data and the regression line for levels 50 to 74
plt.scatter(levels, XP, label='Actual XP (50 to 74)')
plt.plot(levels, predicted_XP_all[:len(levels)], color='red', label='Regression Line (50 to 74)')

# Plot the regression line for levels 75 to 800
plt.plot(all_levels, predicted_XP_all, color='blue', label='Regression Line (75 to 800)')
plt.xlabel('Current Level')
plt.ylabel('XP for Next Level')
plt.title('XP Required for Next Level vs. Current Level')
plt.legend()
plt.grid(True)
plt.show()

# Display the coefficients
print("Coefficient 'a' (slope):", a)
print("Coefficient 'b' (y-intercept):", b)
```
